```python

def extract_side_by_side_tables(image_path):
    # 1. Load and Preprocess
    img = cv2.imread(image_path)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("greay:")
    # plt.figure(figsize=(200, 400))
    plt.imshow(gray)
    plt.show()


    # Thresholding to isolate the table structures (handling the red text)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
    # plt.figure(figsize=(200, 400))
    print("thresh:")
    plt.imshow(thresh)
    plt.show()

    # 2. Find Bounding Boxes for each table
    # We use a vertical kernel to detect columns/table blocks
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (10, 50))

    cnts, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 3. Sort contours by X-coordinate (Left to Right)
    # This addresses your "may not be in order" requirement
    bboxes = [cv2.boundingRect(c) for c in cnts]
    bboxes = sorted(bboxes, key=lambda b: b[0])

    all_tables = []

    # 4. Process each detected table block
    for i, (x, y, w, h) in enumerate(bboxes):
        if x< img.shape[0] and y<img.shape[1]:
            if w < 10 and h < 10: continue # Filter noise

            # Crop the table from the original image
            roi = img[y:y+h, x:x+w]

            # Perform OCR
            # config='--psm 6' assumes a single uniform block of text
            data = pytesseract.image_to_string(roi, config='--psm 6')

            # Clean up data into a list of rows
            rows = [line.strip().split() for line in data.strip().split('\n') if line.strip()]

        if rows:
            df = pd.DataFrame(rows)
            all_tables.append(df)

    # 5. Merge or return list
    # If they all have the same structure, we can concatenate horizontally
    if all_tables:
        final_df = pd.concat(all_tables, axis=1)
        return final_df
    return None
```

```python
img = "/content/Input_image.png"

df=extract_side_by_side_tables(img)
```
